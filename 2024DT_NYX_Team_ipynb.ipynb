```python
import pandas as pd
import pandas_dedupe
```

```python
#import data from csv file
df = pd.read_csv('CAERS_ProductBased.csv')
#delete all entries with "EXEMPTION 4"
df = df[df.PRODUCT != 'EXEMPTION 4']

#fill the "DATA_EVENT" column with "DATE_FDA_FIRST_RECEIVED_REPORT" if is NaN
df.DATE_EVENT.fillna(df.DATE_FDA_FIRST_RECEIVED_REPORT, inplace=True)

#Computer RAM not large enough to run the whole 200K data, select the most recent 50K
df = df.head(50000)
df
```

```python
#Deduplication the repetitive entries
df = pandas_dedupe.dedupe_dataframe(df, ['DATE_EVENT', 'PRODUCT', 'PATIENT_AGE'])
```

```python
max_value = df['cluster id'].max()

#Combine repetitive entries
df_dict = df.to_dict('records')
data_d = {}
for row in df_dict:
  #identify as the cluster if confident
  if int(row['confidence']) > 0.6:
    row_id = int(row['cluster id'])
    data_d[row_id] = dict(row)
  #identify as a new cluster if not confident
  else:
    row_id = max_value + 1
    max_value += 1
    data_d[row_id] = dict(row)
```

```python
#Comvert back to df and delete cluster columns
df = pd.DataFrame.from_dict(data_d, orient='index')
df = df.reset_index()
del df['Unnamed: 12']
del df['cluster id']
del df['confidence']
```

```python
#Add the column for scoring
df.insert(12, 'OUTCOME_LEVEL', 1, True)
```

```python
outcomes = ['serious', 'hospitalization', 'life threatening', 'disability', 'death']
out_count = {}

label = "CASE_OUTCOME"
#Count each outcome frequency
for outcome in outcomes:
  count = df[label].str.count(outcome).sum()
  out_count[outcome] = count

out_count
```

```python
death_value = 15266 # The primary value that we give to death
hospi_value = 235 # The primary value that we give to hospitalization
life_value = hospi_value * (out_count['hospitalization'] / out_count['life threatening'])
disability_value = hospi_value * (out_count['hospitalization'] / out_count['disability'])
serious_value = hospi_value * (out_count['hospitalization'] / out_count['serious'])
other_value = 1 # The primary value that we give to other
```

```python
outcome_val = []

outcome_val.append(('death', death_value))
outcome_val.append(('hospitalization', hospi_value))
outcome_val.append(('life threatening', life_value))
outcome_val.append(('disability', disability_value))
outcome_val.append(('serious', serious_value))

#Sort outcomes by score from low to high, so the higher score can cover the lower
sorted_by_val = sorted(outcome_val, key=lambda tup: tup[1])

sorted_by_val
```

```python
suspect_coeff = 78.77 / 21.23 # Obtained from floridahealth.gov
concomi_coeff = 1
```

```python
#Score each row of data
for outcome in sorted_by_val:
  for index in df.index:
    if outcome[0] in df['CASE_OUTCOME'][index]:
      df['OUTCOME_LEVEL'][index] = outcome[1]
      #Modify the score weight by suspect/concomi
      if df['PRODUCT_TYPE'][index] == 'suspect':
        df['OUTCOME_LEVEL'][index] *= suspect_coeff
      #print(df['OUTCOME_LEVEL'][index])
```

```python
#export cleaned and scored data
df.to_csv("Scored_result.csv")
```

```python
#Sum and sort the dander index by product name
sum_result = df.groupby('PRODUCT')['OUTCOME_LEVEL'].sum().reset_index()
sum_result = sum_result.sort_values(by='OUTCOME_LEVEL', ascending=False)
sum_result.head(10)
```

```python
#Deduplicate the similar names
result_dedepu = pandas_dedupe.dedupe_dataframe(sum_result, ['PRODUCT'])
```

```python
dedup_dict = result_dedepu.to_dict('records')
#dedup_dict = dedup_dict.reverse()
result_d = {}

#Combine product names and their corresponding danger index
for row in dedup_dict:
    row_id = int(row['cluster id'])
    if row_id in result_d:
      sum_val = result_d[row_id]['OUTCOME_LEVEL']
      sum_val += float(row['OUTCOME_LEVEL'])
      result_d[row_id]['OUTCOME_LEVEL'] = sum_val
    else:
      result_d[row_id] = dict(row)
      result_d[row_id]['OUTCOME_LEVEL'] = float(result_d[row_id]['OUTCOME_LEVEL'])

result_d
```

```python
#Convert to df and delete custer columns
final_result = pd.DataFrame.from_dict(result_d, orient='index')
final_result = final_result.sort_values(by='OUTCOME_LEVEL', ascending=False)
final_result = final_result.reset_index()

del final_result['index']
del final_result['cluster id']
del final_result['confidence']
```

```python
from sklearn.preprocessing import MinMaxScaler

#Normalize danger index
scalar = MinMaxScaler()
column_d = final_result['OUTCOME_LEVEL'].values.reshape(-1,1)
normalized_d = scalar.fit_transform(column_d)
final_result['OUTCOME_LEVEL'] = normalized_d
#print(normalized_d)
```
